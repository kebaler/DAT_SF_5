```python
# The "T" value in the Precipitaiton_In field refers to trace
# I'm replacing "T" with 0.001 as it's 1/10th of the previous smallest value (0.01)
sfrides['Precipitation_In'].replace("T",0.001, inplace=True)
```

```python
# Checking to make sure "T" is gone
sfrides.Precipitation_In.unique()
```

```python
sfrides.info()
```

```python
# I'm using LabelEncoder to turn my catagorical features into 0-(k-1) categories
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
# Setting variables for transformed catagorical features to pass to dataframe
Month = le.fit_transform(sfrides.Month)
Day_of_Week = le.fit_transform(sfrides.Day_of_Week)
Hour_of_day = le.fit_transform(sfrides.Hour_of_day)
Start_Station = le.fit_transform(sfrides.Start_Station)
End_Station_Neighborhood = le.fit_transform(sfrides.End_Station_Neighborhood)
Subscription_Type = le.fit_transform(sfrides.Subscription_Type)
Zip_Code = le.fit_transform(sfrides.Zip_Code)
```

```python
# Passing all features, catagorical & continuous, to same dataframe

features = pd.DataFrame({'Month' : Month,
                       'Day_of_Week' : Day_of_Week,
                       'Hour_of_day' : Hour_of_day,
                       'Start_Station' : Start_Station,
                       'Subscription_Type' : Subscription_Type,
                       'Subscription_Type' : Subscription_Type,
                       'Zip_Code' : Zip_Code,
                       'Mean_Temperature_F' : sfrides.Mean_Temperature_F,
                       'Mean_Humidity' : sfrides.Mean_Humidity,
                       'Mean_Wind_Speed_MPH' : sfrides.Mean_Wind_Speed_MPH,
                       'Precipitation_In' : sfrides.Precipitation_In,
                       'Cloud_Cover' : sfrides.Cloud_Cover
                       })
```

```python
# For some reason precipitation is still an object, which was causing problems
features.info()
```

```python
# Setting Precipitation_In to float so that classifiers work
features[['Precipitation_In']] = features[['Precipitation_In']].astype(float)
```

```python
# Double checking data types - we're good
features.info()
```

```python
features['Zip_Code'].hist(bins=20)
```

```python
# Looking at shape of features
features.shape
```

```python
# Creating target dataframe that we're trying to predict
target = pd.DataFrame({'End_Station_Neighborhood' : End_Station_Neighborhood})
```

```python
# This was causing problems, as we need to pass a (n,) sh
target.shape
```

```python
target = target.values.reshape(129850,)
```

```python
target.shape
```

#### Decision Tree

```python
from sklearn import tree
```

```python
# Decision tree with max depth of 10
clf = tree.DecisionTreeClassifier(max_depth=10, random_state=0,)
clf.fit(features.values, target)
```

```python
# Plotting feature importanc
%pylab --no-import-all
numfeat = 11
indices = np.argsort(clf.feature_importances_)[::-1][:numfeat]
bar(xrange(numfeat), clf.feature_importances_[indices], align='center', alpha=.5)
xticks(xrange(numfeat), features.columns[indices], rotation='vertical', fontsize=12)
xlim([-1, numfeat])

ylabel('Feature importances', fontsize=12)

suptitle('Feature importances computed by Decision Tree', fontsize=16)

f = gcf()
subplots_adjust(bottom=0.3)
f.set_size_inches(11, 8);
f.savefig('feature_importance.png', dpi=150);
# credit to http://nbviewer.ipython.org/github/herrfz/dataanalysis/blob/master/assignment2/samsung_data_prediction_submitted.ipynb
```

```python
# printing cross validation score
scores = cross_val_score(clf, features.values, target, cv=10)
print scores
print mean(scores)
```

####PCA

```python
from sklearn.decomposition import PCA
```

```python
features.info()
```

```python
pca = PCA(n_components=2)
X_r = pca.fit(features.values).transform(features.values)
```

```python
ratios = pca.explained_variance_ratio_
print ratios
```

```python
print "First component: " + str(pca.explained_variance_ratio_[0])
print "Second component: " + str(pca.explained_variance_ratio_[1])
#print "Third component: " + str(pca.explained_variance_ratio_[2])
#print "Fourth component: " + str(pca.explained_variance_ratio_[3])
```

```python
print X_r
```

```python
%pylab inline
import matplotlib.pyplot as plt

comp_id = [1, 2]             # id number of component
fig = plt.figure(figsize=(8,5))
plt.plot(comp_id, ratios, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
```

```python
from sklearn.ensemble import RandomForestClassifier

clf5 = RandomForestClassifier(max_depth=10, random_state=0)
clf5.fit(X_r, target)
scores = cross_val_score(clf5, features.values, target, cv=10)
print scores
print mean(scores)
```
